# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Owner\Desktop\Alex_the_Analyst\Customer Call List.xlsx")
```

```python
df
```

```python
# get rid of all duplicates

df = df.drop_duplicates()
```

```python
df
```

```python
# delete the column

df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
# removing extra symbols from the left and right of the last name column

df['Last_Name'] = df['Last_Name'].str.lstrip("...")   
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.rstrip("_")


# aslo removing unwanted symbols

# df["Last_Name"].str.strip("123._/")                 
```

```python
df
```

```python
# removing all unwanted symbols and making all numerical values the same format in the phone numbers column

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True) 
```

```python
df
```

```python
# converting each value to a string so i can...

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# ...add a - inbetween the numbers in standard phone number format

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
```

```python
df
```

```python
# removing any other value in the column I dont want in there and replacing with an empty space, one by one

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
```

```python
df
```

```python
# splitting columns for the heck of it

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n = 2, expand = True)
```

```python
df
```

```python
# replacing values for consistancy in the columns

df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
```

```python
df
```

```python
# getting rid of the selected characters across the whole dataframe and replacing with a blank space

df = df.replace('N/a', '')

# replacing the NaN values with a blank space

df = df.fillna('')
```

```python
df
```

```python
# only selecting the rows that did NOT approved for us to call them if the time comes and removing them from the df

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
```

```python
df
```

```python
# selecting the rows where the Phone_Number value is null and removing them from the df 

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
        
# another way you could drop null values--> df = df.dropna(subset='Phone_Number'), inplace = True
```

```python
df
```

```python
# removing all of the remaining unnecessary columns from the df

df = df.drop(columns = 'Paying Customer')
df = df.drop(columns = 'Address')
df = df.drop(columns = 'Street_Address')
df = df.drop(columns = 'State')
df = df.drop(columns = 'Zip_Code')
```

```python
df
```
